# Conditional examples

## IF Statements

```python
# I'm creating variable "userName" to store the answer to the questions "What's your name?"
userName = input("What is your name? ")
userAge = int(input("What year were you born? "))

# Calculate the users age
age = 2026 - userAge 

print ("Hello " + userName + ",")
print ("You are " + str(age) + " years old.")

# IF statement example
if age >= 21:
    print("Welcome to the bar, enjoy your cold snack!")
```

## Two-Way IF statement example

```python
# I'm creating variable "userName" to store the answer to the questions "What's your name?"
userName = input("What is your name? ")
userAge = int(input("What year were you born? "))

# Calculate the users age
age = 2026 - userAge 

print ("Hello " + userName + ",")
print ("You are " + str(age) + " years old.")

# Two-Way IF example
if age >= 21:
    print("Welcome to the bar, enjoy your cold snack!")
else:
    print("Sorry, no cold snack for you!")
```

## Ternary Expression example

```python
# I'm creating variable "userName" to store the answer to the questions "What's your name?"
userName = input("What is your name? ")
userAge = int(input("What year were you born? "))

# Calculate the users age
age = 2026 - userAge 

print ("Hello " + userName + ",")
print ("You are " + str(age) + " years old.")

# Ternary syntax
canEnter = "You can enter the bar " if age >= 21 else "You cannot enter!"

print(canEnter)
```

## Nested IF statements

```python
# I'm creating variable "userName" to store the answer to the questions "What's your name?"
userName = input("What is your name? ")
userAge = int(input("What year were you born? "))

# Calculate the users age
age = 2026 - userAge 

print ("Hello " + userName + ",")
print ("You are " + str(age) + " years old.")

# Nested IF examples
if age >= 18:
    if age <= 25:
        print("You can enlist in the service.")
    else:
        print("You are too old to enlist.")
else:
    print("You are not eligible for enlistment.")
```

## ELIF Statements
ELIF statements help us check to multiple conditions.

```python
# I'm creating variable "userName" to store the answer to the questions "What's your name?"
userName = input("What is your name? ")
userAge = int(input("What year were you born? "))

# Calculate the users age
age = 2026 - userAge 

print ("Hello " + userName + ",")
print ("You are " + str(age) + " years old.")

# ELIF statement example
if age < 16:
    print("You ar too young to drive.")
elif age >= 16 and age <= 18:
    print("You are eligible for a learner's permit.")
elif age >25 and age <60:
    print("You may be eligible for a insurance discount.")
else:
    print("You are eligible for a driver's license.")
```

## Case Match statements
Useful when you have many different conditions you want to test for

```python
dayToView = int(input("What day number (0 to 6) would you like to show?"))

day = dayToView
day_of_week = ""


if day == 0:
    day_of_week = "Sunday"
elif day == 1:
    day_of_week = "Monday"
elif day == 2:
    day_of_week = "Tuesday"
elif day == 3:
    day_of_week = "Wednesday"
elif day == 4:
    day_of_week = "Thursday"
elif day == 5:
    day_of_week = "Friday"
elif day == 6:
    day_of_week = "Saturday"
else:
    day_of_week = "Invalid day"

print(day_of_week)
```

```python
# Shorthand case match statements
dayToView = int(input("What day number (0 to 6) would you like to show?"))

day = dayToView
day_of_week = ""

case = {
    0: "Sunday",
    1: "Monday",
    2: "Tuesday",
    3: "Wednesday",
    4: "Thursday",
    5: "Friday",
    6: "Saturday"
}

day_of_week = case.get(day, "Invalid day")

print(day_of_week)
```

# ATM Simulator

```python
# Initialize an account balance
balance = 1000


while True:
    print("Welcome to the ATM Simulator")
    print("1. Check Balance")
    print("2. Withdraw")
    print("3. Deposit")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        print("Your current balence is $" + str(balance))
    elif choice == 2:
        amount = int(input("Enter the amount to withdraw: "))
        if amount > balance:
            print("Insufficient funds!")
        elif amount <= 0:
            print ("Invalid amount!")
        else:
            balance -= amount 
            print("$" + str(amount) + " withdrawn successfully!")
            print("Your remaining balance is $" + str(balance))
    elif choice == 3:
        amount = int(input("Enter the amount to deposit: "))
        if amount <= 0:
            print("Invalid amount")
        else:
            balance += amount
            print("$" + str(amount) + " deposited successfully!")
            print("Your remaining balance is $" + str(balance))
    elif choice == 4:
        print("Thank you for using our ATM!")
        break
    else:
        print("Invalid choice, please try again!")
```
