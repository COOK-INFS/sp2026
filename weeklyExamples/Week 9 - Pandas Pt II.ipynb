## Look at the number of passengers per class

```python
class_counts = df['Pclass'].value_counts().sort_index()

class_counts
```

## Odds of surviving by passenger class

```python
# First Group by class and survival
class_survival_counts = df.groupby('Pclass')['Survived'].value_counts().unstack()

# Calculate the ods of survival for each class
odds_of_survival = class_survival_counts[1] / class_survival_counts[0]

# Print the results
odds_of_survival
```

# What was the name and year of the movie with the highest rating and highest number of votes.

```python
# Join the dataframes we need
mvRating = pd.merge(dfTitles, dfRatings, on='tconst')

mvRating
```

```python
# Query to reduce the data to only movies
ratingSort = (mvRating.query('titleType == "movie"'))

ratingSort
```

```python
# Sort our dataframe to finish answering our question
ratingSort.sort_values(by='averageRating', ascending=False)
```

```python
# Sort by multiple columns to accurately answer the question
ratingSort.sort_values(by=['averageRating', 'numVotes'], ascending = False)
```

```python
# Sort by multiple columns to accurately answer the question
ratingSort.sort_values(by=['numVotes', 'averageRating'], ascending = False)
```

### Question:
What was the name and year of the movie with the highest rating and highest number of votes.

```python
# Let's format the output for the end user.
semi1 = (ratingSort.sort_values(by=['averageRating', 'numVotes'], ascending=False))

# Create a dataframe with only the asked for columns.
semi2 = semi1[['primaryTitle', 'startYear', 'averageRating', 'numVotes']]

# Select a subset of the rows for the report
final = (semi2.head(5))

print(final)
```
