# Strings examples

```python
myString = "Hello World"

print(myString)

# Print an index location
print(myString[3])

# Print and index range
print(myString[0:7])
```

## String immutability

```python
string1 = "Bob"
string2 = 'Mary'

# Show the memory location 
memory_before_string1 = id(string1)
memory_before_string2 = id(string2)

# Change the values of our strings
string1 = 'Mary'

# Show the memory location
memory_after_string1 = id(string1)
memory_after_string2 = id(string2)

# Output the memory locations
print(f"Memory location of string1 before change:  {memory_before_string1}")
print(f"Memory location of string2 before change:  {memory_before_string2}")
print(f"Memory location of string1 before change:  {memory_after_string1}")
print(f"Memory location of string2 before change:  {memory_after_string2}")
```

## Formatted Strings

```python
# Concatenating strings
name = "Alice"
age = 30
greeting = "Hello, " + name + ". You are " + str(age) + " years old."

print(greeting)


# Using an "f-string"
```

```python
# F strings
name = "Alice"
age = 30
greeting = f"Hello, {name}. You are {age} years old."

print(greeting)


```

```python
# Using the .format method
name = "Alice"
age = 30
greeting = "Hello, {}. You are {} years old.".format(name, age)

print(greeting)
```

## Escaping characters

```python
escapeEx = "I am a \"goof\", no doubt about it. "

print(escapeEx)
```

```python
# Newline example
newLine = "This is line 1. \nThis is line 2."

print(newLine)
```

```python
# Carriage return vs. newline
carriage = "This is line 1. \rThis is line 2."

print(carriage)
```

## Searching in a string

```python
myString = "I hate always using Hello World."
index = myString.find("World")
print(index)
```

```python
myString = "I hate always using Hello World."
result = "World" in myString
print(result)
```

# Files

## Writing Files

```python
f = open('fileExample.txt', 'w')
f.write("Some new stuff.")
f.close()

```

## Appending


```python
f = open('fileExample.txt', 'a')
f.write("\nEven more stuff.")
f.close()
```

## Reading

```python
f = open('fileExample.txt', 'r')
print(f.read())
f.close()
```

## Using the WITH method

```python
text = "Adding stuff"

with open ('fileExample.txt', 'a') as f:
    f.write(text)
with open ('fileExample.txt', 'r') as f:
    print(f.read())
```

# Error handling

```python
try:
    with open('file2Example.txt', 'r') as f:
        print(f.read())
except FileNotFoundError:
    print("I could not find your file. Please check the filename.")
```
